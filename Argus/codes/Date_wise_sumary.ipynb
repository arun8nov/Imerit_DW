```python
li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    
    # os.path.basename gets 'data.csv'
    # os.path.splitext(...)[0] gets 'data'
    file_name_only = os.path.splitext(os.path.basename(filename))[0]
    
    df['date'] = file_name_only
    li.append(df)

final_df = pd.concat(li, axis=0, ignore_index=True)
```
